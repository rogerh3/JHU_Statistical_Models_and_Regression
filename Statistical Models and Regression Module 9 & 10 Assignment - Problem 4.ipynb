# Calculating PRESS Residuals

```python
# Assuming a basic Linear Regression Model, we can calculate our y-hat and ei values rather easily
# y = b0 + b1x1 + b2x2 + b3x3 + b4x4 + b5x5 + b6x6 + b7x7 + b8x8 + b9x9
# I opted to calculate yhat and ei in the excel file before loading it due to errors trying to loop through rows and calculating

# Assigning Data for Modeling Purposes
y = df[['y']]
x1 = df.iloc[:,1:10]

#Checking x to make sure it is correct
x1
```

```python
# Adding the constant column to x for B0 calculations
x = sm.add_constant(x1) 
x
```

```python
# Creating our model and checking data results 
model = ols('y ~ x', data=df).fit()

est = sm.OLS(y, x).fit() 
est.summary()
```

```python
# Checking to make sure OLS provides similar results to a LM
est2 = LinearRegression()
est2.fit(x1, y)

print(est2.intercept_, est2.coef_, est2.score(x1, y))
```

```python
# Since I previously calculated the yhat and ei in excel using the beta coefficient values from the model above
# we can go into calculating the PRESS Residuals
```

```python
# Assigning ei
ei = df[['ei']]
print(ei)
```

```python
# Calculating the hat matrix (H)
hat = x.dot(np.linalg.inv(x.T.dot(x)).dot(x.T))
hat
```

```python
# Assigning the hat diagonal values
hii = pd.DataFrame(np.diag(hat))
print(hii)
```

```python
PRESS_numerator = pd.DataFrame(np.divide(ei, (1-hii)))
PRESS = pd.DataFrame(np.divide(PRESS_numerator, np.sqrt((np.var(ei)))))
PRESS.rename(columns = {'ei': 'PRESS values'}, inplace = True)
print(PRESS)
```
